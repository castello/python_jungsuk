# 3.  변수란?

- 객체에 붙이는 고유한 이름. 같은 이름에 다른 객체를 연결할 수 있다. 
- scope에서 관리하는 namespace에 (변수이름, 객체)로 저장된다.

### 변수에 타입이 없다. 변수와 연결된 객체에 타입이 있는 것일뿐
- 파이썬은 변수 선언이 없다.
- 타입을 지정할 수 있으나, 주석으로 인식될 뿐이다.

```python
var : int    # 변수 var의 타입이 int라는 것을 알려주는 힌트. 변수의 선언이 아님에 주의
```

```python
try :
    var  # var라는 이름이 없다. 현재 scope의 namespace에 var가 존재하지 않는다.
except Exception as e :
    print(e)
```

### 변수는 반드시 값을 할당해야 정의된다.

- 파이썬에서는 변수 정의할 때 별도의 키워드가 없다.
- 파이썬에는 변수를 저장하고 관리하는 영역인 네임스페이스가 있다.
- 변수 정의는 이 네임스페이스에 변수 이름을 키로 하고 값을 저장하는 구조이다

```python
var = 100   # 변수에 객체를 처음 대입할 때, 변수이름과 객체가 묵여서 (var, 100의 객체주소)으로 namespace에 등록된다. 
```

### 변수에 저장된 값을 사용할 때는 반드시 변수 이름을 사용

- 변수에 대한 저장위치를 찾아서 내부의 값을 반환
- 변수를 참조하는 것은 실제 변수가 저장된 네임스페이스에서 그 변수에 있는 값을 조회하는 것

```python
var
```

### 모듈과 전역 변수 - 주피터 노트북 파일은 하나의 모듈

- 모듈에는 변수를 관리하는 네임스페이스가 만들어진다.
- 모듈에 정의된 변수를 전역변수라고 한다.
- 모듈 내에 정의된 변수는 지역변수
- 모듈과 함수 내에 정의된 변수를 별도의 namespace로 관리


### 모듈 내의 네임스페이스를 불러와서 변수 이름으로 변수의 값을 조회한다

```python
globals()['var'] # 전역 namespace에서 'var'라는 이름과 연결된 객체(100)를 조회
```

## 변수를 식별하는 것은 아주 중요. 각 이름은 서로 구별되게 지어야
### 변수 이름 짓는 규칙 => 명명규칙

- 첫 문자는 문자나 언더스코어(_) - 숫자로 시작하면 안됨
- 두 번째부터는 문자나 언더스코어 또는 숫자 조합 
- 키워드는 변수 이름으로 사용할 수 없다.

## 이름을 지정하는 방법 
- 1. 카멜 표기법(Camel Case) : 카멜 표기법은 낙타표기법이라고도 하며, 낙타의 등의 모습을 닮아 생긴 이름이다.
    예시) count, countMyApple, countMyAppleAndPeach
    
- 2. 스네이크 표기법(Snake Case) : 스네이크 표기법은 한 문자마다 _(underscore)를 붙여 이어나가는 방식
    예시) count, count_my_apple, count_my_apple_and_peach

- 변수의 값을 변경할 때는 할당연산자를 사용해서 새로운 값을 할당할 수 있다.

### 동일한 값 할당할 때
- 변수와 할당 연산자를 여러 개 사용해서 값을 할당

```python
x = y = z = 100
```

```python
x, y, z   # 3개의 값을 하나로 묵어서 객체(튜플)에 저장
```

###  변수를 콤마로 구분해서 처리할 때 주의
- 파이썬은 콤마로 분리하면 튜플로 처리하므로 여러 변수를 지정하고 값을 하나만 할당하면 예외를 발생한다.

```python
try :
   a,b,c = 100 # 에러.
 # a,b,c = 100, 200, 300 # 3개의 값을 할당해야 함
except Exception as e :
    print(e)
```

### 튜를을 만들어서 여러 변수에 값을 나누기

- 튜플은 여러 원소를 가지는 자료구조
- 튜플은 한번 생성되면 변경이 불가능하다.(readonly, immutable)

```python
t = tuple([1,2,3])   # tuple()을 이용해서 list인 [1,2,3]을 튜플로 변환
```

```python
a,b,c = t   # 튜플 t를 a,b,c에 나눠서 저장
```

```python
print(a)
print(b)
print(c)
```

## 연산자 알아보기

- 연산자를 사용하지만,  연산자는 내부적으로 메소드로 처리된다.

```python
aa = 100; bb = 200; # =는 연산자가 아니라 대입문. 파이썬에 대입 연산자는 없음.
```

```python
aa ** 2 ; bb ** 2 #  aa.__pow__(2)로 처리됨. **연산자는 __pow__메서드로 처리
```

```python
aa + bb # +연산자는 __add__메서드. aa.__add__(bb), bb.__radd__(aa)와 동일
```

```python
aa.__add__(bb)
```

### 나누기는 2개 제공

- 나머지까지 처리하는 경우(__truediv__)와 나머지를 제외한 처리(__floordiv__)

```python
aa / bb, aa//bb #  //는 나머지없이 몫만 반환
```

```python
aa.__truediv__(bb), aa.__floordiv__(bb)
```

```python
aa % bb   # 나머지를 반환하는 연산자 %는 __mod__메서드
```

### 정수를 실수로 변환해서 처리

```python
cc = float(aa); dd = float(bb)  # float()는 생성자. 형변환에 사용
```

```python
cc / dd, cc//dd
```
