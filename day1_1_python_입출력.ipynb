## 2.2 f-string 권장 - 템플릿 + __ format__
- f-string은 결국 format() 호출, 그 핵심 훅(hook)이 바로 __ format__
- 대부분 __str__, __repr__로 충분. __format__은 잘 안씀.(형식화에 사용)

```python
x = 10
print(f"x = {x}")    # x = 10
print(f"{x=}")       # x=10   print(format(x,"x="))과 동일
```

```python
print(f"{3.14159:.2f}")     # float.__format__ 소수점 아래 2자리
print(f"{10:08b}")          # int.__format__  10을 8자리 이진수로, o, x
# print(f"{dt:%Y-%m-%d}")     # datetime.__format__
# 이게 가능한 이유 = 각 타입이 __format__ 구현
```

## 2.3 logging 모듈 - stderr에 출력
- print대체. 실무에서 필수.
- 이미 다른 곳에 logging 설정이 되어 있으면, basicConfig()는 아무것도 안 함

#기존의 설정을 무시하고, 강제 설명
logging.basicConfig(level=logging.INFO,force=True) 

logging.info("msg")  
       ↓  
root logger  
       ↓ (level >= INFO?)  
StreamHandler  
       ↓  
Formatter  
       ↓  
sys.stderr.write()  

```python
import logging

# 루트 로거(root logger)를 초기화. 레벨이 INFO 이상인 로그만 처리하도록 설정
# DEBUG < INFO < WARNING < ERROR < CRITICAL
logging.basicConfig(level=logging.INFO)
logging.info("info")    # INFO:root:info 콘솔 출력시
logging.debug("debug")  # DEBUG는 INFO보다 아래라서 출력x
logging.error("error")  # ERROR:root:error
```

```python
import logging

# 실무에서 많이 사용되는 출력 형식
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(name)s: %(message)s",
    force=True
)
logging.info("hello")
logging.getLogger().info("hello") # 위와 동일 getLogger()가 root logger반환
```

```python
print("이름을 입력해주세요:")
name = input()
name
```

```python
# 안전한 숫자 입력
while True:
    print("숫자를 입력해주세요:")
    try:
        n = int(input())
        break
    except ValueError:
        print("retry")
print(n)
```

```python
# jupyter notebook에서는 ctrl+D인식이 잘안됨. 콘솔에서 실행
inputs = []
while True:
    try:
        line = input() # 여러줄 입력받다 ctrl+D누르면 종료
        inputs.append(line)
        print(f"line={line}")
    except EOFError:
        break
    
print(inputs)
```

### 3.5 argparse
- sys.argv는 “원재료”, argparse는 “조리된 완성품”
- sys.argv 위에 올라간 프레임워크. 대부분 사용
- sys.argv를 읽어서 파싱
- 타입 변환 자동, 기본값, 필수 옵션
- 검증 + 에러 메시지 제공
- --help 자동 생성  
  
OS  
 ↓  
argv 문자열 배열  
 ↓  
sys.argv   ← raw  
 ↓  
argparse.parse_args()  
 ↓  
Namespace(count=3)  

```python
# 편리
import argparse

parser = argparse.ArgumentParser()
parser.add_argument("count", type=int)
args = parser.parse_args()

print(args.count)

# 에러시 자동으로 아래와 같은 메시지 출력
# usage: app.py [-h] --count COUNT
# app.py: error: argument --count: invalid int value: 'abc'

# sys.argv는 아래와 같이 출력
# ValueError: invalid literal
```
